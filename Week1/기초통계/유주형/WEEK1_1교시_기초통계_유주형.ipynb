## A1. 답안 작성  
답변:
1. A/B 테스트란?

A/B 테스트는 같은 목표(전환, 클릭, 구매, 체류 시간 등)를 두고, 서로 다른 버전 (A와 B)를 사용자에게 무작위로 노출한 뒤 성과 차이가 우연인지(노이즈) 실제 효과인지를 통계적으로 판단하는 실험을 말한다.

다시말해, 실제 사용자 행동 데이터를 이용한 의사 결정의 표준 도구이다. 따라서 핵심은 '무작위 배정', '동일 조건', '사전 정의된 지표'가 된다.

A(Control): 현재 버전, 기준

B(Alternative): 바꾸고 싶은 새로운 버전(문구, 버튼 색, 가격 표시, 레이아웃, 추천 순서 등)

2. A/B 테스트의 예시

상황: 암/노화 연구실에서 신규 논문을 발표했고, 공동 연구 제안, 학생/포닥 지원, 데이터 요청을 늘리고 싶다.

가설: "논문의 biological significance를 먼저 강조하면, 연구자들의 문의 전환률이 증가할 것이다.

2-1. 실험 설계

A: 홈페이지 상단의 기술 중심 설명(We identified a novel mTORC1-FAM120A axis regulating SASP)

B: 홈페이지 상단의 임상적 의미 제시(Targeting mTORC1-FAM120A rewires SASP and alters tumor-stroma communication)

2-2. 핵심 지표 (KPI)

+ 보조 지표: 페이지 체류 시간, PDF 다운로드 후 이탈률

2-3. 유의 사항

2-3-1. 초두 효과

> 초두 효과(Primacy effect): 처음 본 경험이 이후 판단을 강하게 좌우하는 현상이다. 기존 방식에 익숙한 유저는 새 방식을 불편하다고 느껴 성과가 떨어질 수 있고, 반대로 신규 유저는 새 방식이 더 직관적일 수 있다. 즉, 전체 평균만 보면 효과가 왜곡될 수 있음을 고려해야 한다.

> 신규 유저 vs 기존 유저 분리 분석(cohort split), 런인(run-in) 기간, 적응 기간 고려, 교차 오염 방지(sticky assignment) 등을 활용해서 해결할 수 있다.

연구자는 첫 문단에서 논문의 급이나 핵심 내용을 판단하는 경우가 많고, 연구실의 기존 연구 방향에 익숙한 사람은 mechanistic detail, molecular mechanism을 선호하지만, 신규 분야 연구자나 MD phd는 biological/clinical relevance를 선호하는 경향이 있다. 따라서 분리 분석이 필요하다.

2-3-2. 표본 수 & 실험 기간

연구실 홈페이지의 특성상 하루 방문자 수가 적고, noise가 커진다. 반면, 학회 시즌이나 논문 발표 직후에는 트래픽이 급증하기 때문에 confounder가 발생한다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:

p1, p2는 두 그룹의 전환율로 잡을 수 있다(e.g. 기존 CVR이 25% 라면  p1 = 0.25). 이 둘의 차이 p1-p2가 검출하고 싶은 효과 크기(MDE)이다.

p(1-p)는 비율의 분산이 된다. 베르누이 시행에서의 분산이며, 전환 여부가 성공/실패의 이진 변수이기 때문이다.

분모 (p1-p2)^2 은 효과 크기의 제곱으로, 차이가 작을 수록 필요한 표본 수가 급증하게 된다는 직관과 일치한다.

결과로 나온 n은 각 그룹 당 필요한 표본 개수를 의미한다. (총 필요 표본 수는 2n)

유의 수준이 α = 0.05, 검정력을 80%로 가정했으므로 코드는 다음과 같다.

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.

import math
from scipy.stats import norm

def sample_size(p1, p2, alpha=0.05, power=0.8):

  # z 값 계산
  z_alpha_over_2 = norm.ppf(1 - alpha/2)
  z_beta = norm.ppf(power)

  # 분산 항
  var = p1 * (1 - p1) + p2 * (1 - p2)

  # 효과 크기
  effect_size = (p1 - p2) ** 2

  # 표본 수 계산
  n = ((z_alpha_over_2 + z_beta) ** 2 * var) / effect_size


  # 결과 출력(표본 개수는 올림 처리)
  return math.ceil(n)
```

```python
# 예시 데이터
p1 = 0.25 # 기존안의 CVR
p2 = 0.30 # 변경안의 CVR
n = sample_size(p1, p2)
print("총 필요 표본 수(minimum sample size, n):", n)
```

이유는 본 실험을 메인 홈 검색 모듈의 위치와 클릭 여부라는 두 범주형 변수 간의 독립성 여부를 검정하는 문제로 볼 때 카이 제곱을 이용한 독립성 검정을 적용할 수 있기 때문이다.

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

## A3-3. 아래 검정 이름과 이유를 채워주세요.
답변: 위 가설에 필요한 검정은 one-sided z-test 이고,  

이유는 카드 클릭 여부는 클릭(1), 미클릭(0)의 이진 변수이며, CTR은 이진 변수의 비율이 된다. 표본 수가 충분히 크면 표본 비율은 정규 분포로 근사 가능하며(CLT) chi-square test와 달리 모비율의 비율 차이를 검정하는 것이 가능하기 때문이다.

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성

import math
from scipy.stats import norm

# 가정 데이터 (CTR)

# 메인 페이지 방문 수
nA = 5000   # 기존안 노출 수
nB = 5000   # 변경안 노출 수

# 카드 클릭 수
clickA = 600   # CTR_A = 0.12
clickB = 750   # CTR_B = 0.15

alpha = 0.05  # 유의수준

# 두 모비율 Z-test (단측)
# H0: pA = pB
# H1: pB > pA

pA_hat = clickA / nA
pB_hat = clickB / nB

# pooled proportion (H0 가정)
p_pool = (clickA + clickB) / (nA + nB)

# 표준오차
se = math.sqrt(p_pool * (1 - p_pool) * (1/nA + 1/nB))

# Z 통계량
z = (pB_hat - pA_hat) / se

# 단측 p-value
p_value = 1 - norm.cdf(z)

print(f"A CTR = {pA_hat:.4f} ({pA_hat*100:.2f}%)")
print(f"B CTR = {pB_hat:.4f} ({pB_hat*100:.2f}%)")
print(f"z = {z:.4f}")
print(f"p-value = {p_value:.6f}")

if p_value < alpha:
    print("결론: 귀무가설 기각 → 변경안의 CTR이 유의하게 높다")
else:
    print("결론: 귀무가설 기각 실패 → CTR 증가가 유의하지 않다")
```

**결과해석**

답변: 두 모비율에 대한 one-sided z-test 결과, p-value가 유의수준 0.05보다 작아 귀무 가설을 기각하였다. 이는 메인 화면에서 결정적인 공모전 2-3개를 우선 노출하는 변경안이 기존안 대비 카드 클릭률을 유의하게 증가시킨다는 근거가 된다.
